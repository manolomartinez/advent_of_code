```python
from z3 import *
```

```python
with open("input.txt") as f:
    lines = f.readlines()
```

```python
def parse(line):
    line = line.split(' ')
    buttons = line[1:-1]
    joltages = line[-1]
    buttons = [[eval(number) for number in button[1:-1].split(',')] for button in buttons]
    joltages = [eval(number) for number in joltages[1:-2].split(',')]
    return buttons, joltages
machines = [parse(line) for line in lines]
```

```python
def z3solver(machine):
    buttons, joltages = machine
    x = [Int('x%s' % i) for i in range(len(buttons))]
    lhs = [[int(j in b) for b in buttons] for j in range(len(joltages))]
    o = Optimize()
    for b in range(len(buttons)):
        o.add(x[b]>=0)
    for j in range(len(joltages)):
        o.add(sum(x[b] for b in range(len(buttons)) if lhs[j][b])==joltages[j])
    o.minimize(sum(x))
    o.check()
    return sum(o.model()[xval].as_long() for xval in x)
```

```python
sum(z3solver(machine) for machine in machines)
```
